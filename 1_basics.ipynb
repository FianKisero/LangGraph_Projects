# Jupyter Notebook Basics

This notebook provides a diagnostic and learning guide for opening and working with Jupyter notebooks in VS Code.

## 1. Check File Existence and Path

Verify that `1_basics.ipynb` exists in your workspace folder.

```python
import os
import pathlib

# Check if the notebook file exists
notebook_path = pathlib.Path("1_basics.ipynb")
print(f"Notebook exists: {notebook_path.exists()}")
print(f"Absolute path: {notebook_path.resolve()}")

# List all .ipynb files in the current directory
ipynb_files = list(pathlib.Path(".").glob("*.ipynb"))
print(f"\nNotebook files found: {ipynb_files}")
```

## 2. Verify Jupyter and Python Extensions

Ensure the VS Code Jupyter and Python extensions are installed:
- **ms-toolsai.jupyter** (Jupyter)
- **ms-python.python** (Python)

To check:
1. Go to Extensions (Ctrl+Shift+X)
2. Search for "Jupyter" and "Python"
3. Verify they are installed and enabled
4. Update them if an update is available

## 3. Open the Notebook in VS Code

To open a notebook file:
1. Go to **File > Open File** (Ctrl+O)
2. Navigate to `1_basics.ipynb` in the workspace
3. The file should open in **Notebook Editor** (not plain text)

If it opens in plain text, right-click the file tab and select "Reopen with..." → "Jupyter: Notebook"

## 4. Select Python Interpreter and Kernel

To choose the correct Python environment:
1. Click the **kernel selector** in the top-right (shows current kernel)
2. Select **"Python 3.12.7 (.venv)"** or your preferred environment
3. The kernel must match your virtual environment

If no kernel appears:
- Activate your virtual environment first: `.\.venv\Scripts\Activate.ps1`
- Install jupyter: `pip install jupyter`

## 5. Check File Permissions and Workspace Settings

**Workspace Trust:**
- If prompted "Do you trust the authors of the files in this folder?", click **Trust**
- Otherwise go to **File > Trust this Workspace**

**File Permissions:**
- Ensure you have write permissions on the notebook file
- Right-click the file in Explorer → Properties → Check "Read-only" is unchecked

## 6. Restart VS Code and Reload Window

If issues persist:
1. Press **Ctrl+Shift+P** and run **"Developer: Reload Window"**
2. Or completely restart VS Code
3. This applies extension updates and recovers from transient issues
